```python
from quantfolio import Portfolio
from quantfolio.Optimizers import bayesianOptimizer
```

```python
# Download financial data from Yahoo Finance and store them in the data folder

p = Portfolio()
p.addAsset("../data/A35.csv", "A35")
p.addAsset("../data/ES3.csv", "ES3")
p.addAsset("../data/IWDA.csv", "IWDA")

# Add exchange rate data (also downloaded from Yahoo Finance)
p.addExchangeRate("../data/forex/USDSGD.csv", "USD", False)

# Adjust foreign currency denominated assets to SGD 
p.exchangeAdjustment(2, "USD")
```

```python
bayesianRes = bayesianOptimizer(p, sims=50)
```

```python
bayesianRes.plotEfficientFrontier()
```

```python
bayesianRes.plotConvergence()
```
